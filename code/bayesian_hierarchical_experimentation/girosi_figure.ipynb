```python
import pandas as pd
import numpy as np
```

```python
dl_df = pd.read_csv('mses_by_country.csv', index_col=0)
```

```python
print(dl_df.to_string())
```

```python
dl_df['Percent Improvement'] = ((dl_df['Lee-Carter'] - dl_df['Deep Learning']) / dl_df['Lee-Carter']) * 100
```

```python
print(dl_df.to_string())
```

```python
dl_df['Girosi and King Percent Improvement'] = np.nan
```

```python
print(dl_df.to_string())
```

```python
values_from_book = {
    'CAN': 45.1, 'USA': 27.9, 'ISR': 37.7, 'JPN': 33.6, 'AUT': 33.6, 'BEL': 22.4, 'BGR': 1.2, 
    'DNK': 37.8, 'FIN': 30.6, 'FRATNP': 19.6, 'GRC': 11.9, 'HUN': -19, 'ISL': 30, 'IRL': 62.2, 
    'ITA': 41.2, 'LUX': 15.1, 'NLD': 28.4, 'NOR': 49.1, 'POL': -15.2, 'PRT': 9.2, 'ESP': 15.8, 
    'SWE': 30.5, 'CHE': 22.9, 'AUS': 20.9, 'NZL_NM': 36.6
}
```

```python
print(values_from_book)
```

```python
for idx, value in values_from_book.items():
    dl_df.loc[idx, 'Girosi and King Percent Improvement'] = value
```

```python
dl_df = dl_df.dropna()
```

```python
print(dl_df.to_string())
```

```python
final_df = dl_df.iloc[:, -2:].round(1)
final_df['Difference'] = dl_df['Percent Improvement'] - dl_df['Girosi and King Percent Improvement']
final_df = final_df.round(1)
final_df
```

```python
ave = final_df.mean()
med = final_df.median()
final_df.loc['Average'] = ave
final_df.loc['Median'] = med
final_df = final_df.round(1)
final_df
```

```python
latex_code = final_df.to_latex(index=True)
print(latex_code)
```
